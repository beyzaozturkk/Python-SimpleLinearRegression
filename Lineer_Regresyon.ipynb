**LİNEER REGRESYON**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=np.random.rand(100,1)
y=3+2*x+0.5*np.random.rand(100,1)
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x,y)
```

```python
plt.figure()
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue',alpha=0.7)
plt.title('Lineer Regresyon')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
a0=lin_reg.intercept_
a1=lin_reg.coef_
print("intercept:",a0)
print("coefficient:",a1)
```

**diabetes veri seti kullanarak lineer regresyon uygulaması**

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
diabetes_X, diabetes_y= load_diabetes(return_X_y= True)
```

```python
#yeni bir axis oluşturup 2.sütunun tüm satırlarını al
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
diabetes_X_train=diabetes_X[:-20] # son tanesi test veri seti
diabetes_X_test=diabetes_X[-20:]
```

```python
diabetes_y_train=diabetes_y[:-20] # son tanesi test veri seti
diabetes_y_test=diabetes_y[-20:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lin_reg1=LinearRegression()
```

```python
lin_reg1.fit(diabetes_X_train,diabetes_y_train)
```

```python
y_pred=lin_reg1.predict(diabetes_X_test)
```

```python
mse=mean_squared_error(diabetes_y_test,y_pred)
print("MSE:",mse)
r2=r2_score(diabetes_y_test,y_pred)
print("R2:",r2)
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test,color='darkblue')
plt.plot(diabetes_X_test,y_pred,color='black')
```
